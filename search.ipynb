```python
class Node:
    def __init__(self, key=None, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
        
    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node, level=0):
        if node is not None:
            self._printTree(node.right, level + 1)
            print(' '*level*2 + '(' + str(node.key) + ',' + str(node.value) + ')')
            self._printTree(node.left, level + 1)
            
    def __getitem__(self, key):
        if self.root is not None:
            return self.search(key, self.root)
        else:
            return None  
            
    def search(self, key, node):
        if node.key == key:                         # return the value
            return node.value
        elif key < node.key:                        
            if node.left != None:                   # recursive on the left child
                return self.search(key, node.left)
            else:                                  
                return None
        elif key > node.key:                        # recursive on the right child
            if node.right != None:
                return self.search(key, node.right)
            else:
                return None
        return None                                 # base case
```

# Case Example #

```python
Tree = BST()
Tree.root = Node(2, 'b')
Tree.root.left = Node(1, 'a')
Tree.root.right = Node(3, 'c')
Tree.printTree()
print(Tree[1])
print(Tree[2])
print(Tree[3])
print(Tree[4])
```

```python
Tree = BST()
Tree.root = Node(1, 'a')
Tree.root.right = Node(2, 'b')
Tree.root.right.right = Node(3, 'c')
Tree.printTree()
print(Tree[1])
print(Tree[2])
print(Tree[3])
print(Tree[4])
```
