```python
class Node:
    def __init__(self, key=None, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
        
    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node, level=0):
        if node is not None:
            self._printTree(node.right, level + 1)
            print(' '*level*2 + '(' + str(node.key) + ',' + str(node.value) + ')')
            self._printTree(node.left, level + 1)
            
    def __setitem__(self, key, value):
        if self.root is None:
            self.root = Node(key, value)
        else:
            self.insert(key, value, self.root)

    def insert(self, key, value, node: Node):
        if key == node.key:                             # replace new key and value
            node.key, node.value = key, value
            
        elif key < node.key:
            if node.left == None:                       # add new node at the left child
                node.left= Node(key, value)
            else:                                       # recursive on the left child
                self.insert(key, value, node.left)

        elif key > node.key:
            if node.right == None:                      # add new node at the right child
                node.right = Node(key, value)
            else:                                       # recursive on the right child
                self.insert(key, value, node.right)
            
    def maxValueNode(self, node):
        current = node
        while(current.right is not None):
            current = current.right
        return current
        
    def __delitem__(self, key):
        self.root = self.delete(key, self.root)
            
    def delete(self, key, node: Node):
        if node.key == key:                                     # It's a key
            if node.left != None and node.right == None:            # There is only child on left
                node = node.left
                return node
            
            elif node.left == None and node.right != None:          # There is only child on right
                node = node.right
                return node
            
            elif node.left != None and node.right != None:          # There are childs on left and right
                MaxKey:Node = self.maxValueNode(node.left)              # Find the maximum key but smaller than
                                                                        # the key on the left (only for this problem)
                node.left = self.delete(MaxKey.key, node.left)          # recursive to replace that maximum key
                node.key, node.value = MaxKey.key, MaxKey.value         # replace the key and value of the maximum key
                
            else:
                node = None                                         # There is not any childs
            
        else:                                                   # It's not the key
            if node.left != None and node.right == None:            # There is only child on left
                node.left = self.delete(key, node.left)                 # recursive on the left
                
            elif node.left == None and node.right != None:          # There is only child on right
                node.right = self.delete(key, node.right)               # recursive on the right
                
            elif node.left != None and node.right != None:          # There are childs on left and right
                if key > node.key:                                      # The key is greater than the current one
                    node.right = self.delete(key, node.right)               # recursive on the right
                    
                else:                                                   # The key is smaller than the current one
                    node.left = self.delete(key, node.left)                 # recursive on the left

        return node                                             # return node
```
